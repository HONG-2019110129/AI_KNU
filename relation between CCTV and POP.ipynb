```python
CCTV_Seoul = pd.read_csv(cctv_csvPath)
CCTV_Seoul
```

```python
#방법 1
CCTV_Seoul.columns = ['구별', '소계', '2013년도 이전', '2014년', '2015년', '2016년']
```

```python
#방법 2
CCTV_Seoul.rename(columns = {CCTV_Seoul.columns[0] : '구별'}, inplace = True)
```

```python
CCTV_Seoul
```

```python
# 엑셀 파일 불러오기
pop_Seoul = pd.read_excel(pop_xlsPath)
pop_Seoul
```

```python
#세부적 엑셀파일 불러오기(골라서 불러오기) csv는 drop으로 데이터를 가저온후 골라야 되는데 엑셀은 처음부터 가져올때 걸러서 가져올수있다.
pop_Seoul = pd.read_excel(pop_xlsPath,header = 2,usecols='B,D,G,J,N')
pop_Seoul
```

```python
data_result.set_index(keys=['구별'],inplace =True)
```

```python
plt.figure(figsize = (10,8))
data_result['소계'].sort_values().plot(kind ='barh',grid=True)
plt.show()
```

```python
data_result['CCTV 비율'] = data_result['소계']/ data_result['인구수'] * 100
data_result['CCTV 비율'].sort_values().plot(kind = 'barh',grid = True,figsize = (10,8))
```

```python
plt.figure(figsize = (10,10))
plt.scatter(data_result['인구수'],data_result['소계'])
plt.xlabel = ('pop')
plt.ylabel = ('CCTV')
plt.grid()

plt.show()
```

```python
# numpy의 polyfit

fp1 = np.polyfit(data_result['인구수'],data_result['소계'],1)
f1=np.poly1d(fp1)
fx=np.linspace(100000,700000,100)

plt.figure(figsize = (10,10))
plt.scatter(data_result['인구수'],data_result['소계'])
plt.plot(fx,f1(fx), ls ='dashed',lw=3,color = 'g')
plt.grid()

plt.show()
```
